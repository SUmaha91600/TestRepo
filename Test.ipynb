```python
# Store longitude, latitude and street crossing name of each public library location.
class XPoint(object):
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __str__(self):
        return "P(%g_%g)" % (self.x, self.y)

class NamedPoint(XPoint):
    def __init__(self, name, x, y):
        super().__init__(x, y)
        self.name = name

    def __str__(self):
        return self.name

    def __hash__(self):  # Permet d'utiliser set(libraries)
        return hash((self.x, self.y, self.name))

    def __eq__(self, other):
        return isinstance(other, NamedPoint) and self.x == other.x and self.y == other.y and self.name == other.name
        
```

```python
# Simple distance computation between 2 locations.
from geopy.distance import great_circle
 
def get_distance(p1, p2):
    return great_circle((p1.y, p1.x), (p2.y, p2.x)).miles
    
```

```python
def build_libraries_from_url(url, name_pos=8, lat_long_pos=17):
    import requests
    import json

    # R√©cup√©ration des donn√©es depuis l'API
    r = requests.get(url)
    myjson = json.loads(r.text)

    # V√©rifier la pr√©sence de la cl√© "data"
    if "data" not in myjson:
        raise ValueError("Erreur : la cl√© 'data' est absente dans la r√©ponse JSON.")

    myjson = myjson["data"]

    libraries = []
    k = 1
    for location in myjson:
        try:
            # Extraire le nom de la biblioth√®que
            uname = location[name_pos] if name_pos < len(location) else "Unknown"

            # Extraire les coordonn√©es latitude et longitude
            lat_long = location[lat_long_pos] if lat_long_pos < len(location) else None
            if isinstance(lat_long, list) and len(lat_long) >= 3:
                latitude = float(lat_long[1]) if lat_long[1] else None
                longitude = float(lat_long[2]) if lat_long[2] else None
            else:
                latitude = longitude = None

        except (TypeError, ValueError, IndexError):
            latitude = longitude = None
        
        # Ajouter uniquement les biblioth√®ques avec des coordonn√©es valides
        if latitude is not None and longitude is not None:
            name = f"P_{uname}_{k}"
            cp = NamedPoint(name, longitude, latitude)
            libraries.append(cp)
            k += 1

    return libraries

# Ex√©cuter avec les bons indices
libraries = build_libraries_from_url(
    "https://data.cityofchicago.org/api/views/x8fc-8rcq/rows.json?accessType=DOWNLOAD"
)

print(f"There are {len(libraries)} public libraries in Chicago")

# V√©rification des donn√©es extraites
if libraries:
    print(f"Exemple de biblioth√®que : {libraries[0]}")
else:
    raise ValueError("‚ùå La liste des biblioth√®ques est vide. V√©rifiez la structure des donn√©es API.")
```

```python
from docplex.mp.model import Model

mdl = Model("coffee shops")
```

```python

print(f"üìå V√©rification des √©l√©ments dans libraries : {type(libraries)} - {len(libraries)} √©l√©ments")
print(f"üìå V√©rification des √©l√©ments dans coffeeshop_locations : {type(coffeeshop_locations)} - {len(coffeeshop_locations)} √©l√©ments")


BIGNUM = 999999999

# Supprime les doublons correctement
libraries = list({(lib.x, lib.y, lib.name): lib for lib in libraries}.values())
coffeeshop_locations = list(libraries)

# V√©rifier qu'on a bien des √©l√©ments
if not libraries:
    raise ValueError("‚ùå ERREUR: 'libraries' est vide apr√®s suppression des doublons.")

if not coffeeshop_locations:
    raise ValueError("‚ùå ERREUR: 'coffeeshop_locations' est vide apr√®s suppression des doublons.")

# Cr√©ation des variables de d√©cision
coffeeshop_vars = mdl.binary_var_dict(coffeeshop_locations, name="is_coffeeshop")
link_vars = mdl.binary_var_matrix(coffeeshop_locations, libraries, "link")
```

```python
for c_loc in coffeeshop_locations:
    for b in libraries:
        if get_distance(c_loc, b) >= BIGNUM:
            mdl.add_constraint(link_vars[c_loc, b] == 0, "ct_forbid_{0!s}_{1!s}".format(c_loc, b))
```

```python
mdl.add_constraints(link_vars[c_loc, b] <= coffeeshop_vars[c_loc]
                   for b in libraries
                   for c_loc in coffeeshop_locations)
mdl.print_information()
```

```python
mdl.add_constraints(mdl.sum(link_vars[c_loc, b] for c_loc in coffeeshop_locations) == 1
                   for b in libraries)
mdl.print_information()
```

```python
# Total nb of open coffee shops
mdl.add_constraint(mdl.sum(coffeeshop_vars[c_loc] for c_loc in coffeeshop_locations) == nb_shops)

# Print model information
mdl.print_information()
```

```python
# Minimize total distance from points to hubs
total_distance = mdl.sum(link_vars[c_loc, b] * get_distance(c_loc, b) for c_loc in coffeeshop_locations for b in libraries)
mdl.minimize(total_distance)
```
